```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

##################################
outf = r'test_torque_out_0.001dt.txt'
n = 1000
nv = 12

angle = 0
omega = 1
ix = 2
iy = 3
iz = 4
jx = 5
jy = 6
jz = 7
kx = 8
ky = 9
kz = 10
t = nv-1
##################################

f = open(outf, "r")
allLines = f.readlines()



data = np.zeros((n, nv))

for i in range(n):
    datum = allLines[i].split()
    for j in range(nv):
        data[i,j] = float(datum[j])
        
fig, axs = plt.subplots(3,3)
fig.set_size_inches(20, 20)

for i in range(3):
    for j in range(3):
        axs[i,j].set_xlabel('time (yrs)')

axs[0,0].set_ylabel('ix')
axs[1,0].set_ylabel('iy')
axs[2,0].set_ylabel('iz')
axs[0,1].set_ylabel('jx')
axs[1,1].set_ylabel('jy')
axs[2,1].set_ylabel('jz')
axs[0,2].set_ylabel('kx')
axs[1,2].set_ylabel('ky')
axs[2,2].set_ylabel('kz')

axs[0,0].plot(data[:,t], data[:,ix])
axs[1,0].plot(data[:,t], data[:,iy])
axs[2,0].plot(data[:,t], data[:,iz])
axs[0,1].plot(data[:,t], data[:,jx])
axs[1,1].plot(data[:,t], data[:,jy])
axs[2,1].plot(data[:,t], data[:,jz])
axs[0,2].plot(data[:,t], data[:,kx])
axs[1,2].plot(data[:,t], data[:,ky])
axs[2,2].plot(data[:,t], data[:,kz])
```

```python
# plot spin rate
plt.plot(data[:,t], data[:,omega])
```
