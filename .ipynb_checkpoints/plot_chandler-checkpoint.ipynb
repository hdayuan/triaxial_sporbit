```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

##################################
outf = r'chandler_rk_out.txt'
n = 1000
nv = 4

si = 0
sj = 1
sk = 2
t = 3
##################################

f = open(outf, "r")
allLines = f.readlines()

data = np.zeros((n, nv))

for i in range(n):
    datum = allLines[i].split()
    for j in range(nv):
        data[i,j] = float(datum[j])
        
fig, axs = plt.subplots(3,1)
fig.set_size_inches(10, 20)

for i in range(3):
    axs[i].set_xlabel('time (yrs)')

axs[0].set_ylabel('si')
axs[1].set_ylabel('sj')
axs[2].set_ylabel('sk')

axs[0].plot(data[:,t], data[:,si])
axs[1].plot(data[:,t], data[:,sj])
axs[2].plot(data[:,t], data[:,sk])
```
